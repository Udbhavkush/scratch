```python
import numpy as np
import matplotlib.pyplot as plt
import random
from micrograd.nn import MLP
%matplotlib inline
```

```python
np.random.seed(444)
random.seed(444)
```

```python
xs = [
    [-1.0, 2.0, 5.0],
    [4.0, -3.0, 0.5],
    [7.0, 2.0, -3.5],
    [0.0, 6.0, 7.0]
]
yout = [1, -1, -1, 1]

model = MLP(3, [4, 4, 1])
```

```python
steps = 100
lr = 0.01

for k in range(steps):
    # forward pass
    ypred = [model(x) for x in xs]
    # MSE
    loss = sum(((pred - target)**2 for target, pred in zip(yout, ypred)))
    print(loss)
    
    # backward pass
    model.zero_grad()
    loss.backward()
    
    # update parameters
    for p in model.parameters():
        p.data += -lr * p.grad
    
    print(f"{k}, loss: {loss.data}")  
```

```python
y_final_pred = [model(x) for x in xs]
print(y_final_pred)
```
